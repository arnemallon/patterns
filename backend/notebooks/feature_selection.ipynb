### Feature Selection Using Fisher Score

```python
import pandas as pd
import numpy as np

# Read the CSV file using pandas
data = pd.read_csv("/Users/arnemallon/Developer/Datasets/BABD-13.csv")

print("Shape of the dataset:", data.shape)
print("\nFirst 5 rows of the dataset:")
print(data.head())
```

```python
# Prepare the data
# Remove non-numeric columns and the label column
X = data.drop(['label'], axis=1)
y = data['label']

# Convert to numpy arrays
X = X.to_numpy()
y = y.to_numpy()

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=40)

print(X_train)
print(y_train)
```

```python
from sklearn.feature_selection import chi2

# Convert string values to float, replacing non-numeric values with NaN
X_train = pd.DataFrame(X_train).apply(pd.to_numeric, errors='coerce').fillna(0).to_numpy()

# Ensure all values are non-negative by taking absolute values
X_train = np.abs(X_train)

f_p_values = chi2(X_train, y_train)

f_p_values
```

```python
# Get the indices of the sorted chi2 scores
indices = np.argsort(f_p_values[0])[::-1]  # Sort in descending order
chi_scores = f_p_values[0][indices]

# Print top 20 features with their chi2 scores
print("Top 20 features by chi2 score:")
for i in range(20):
    print(f"Feature {indices[i]}: chi2 score = {chi_scores[i]:.2e}")
```

```python
# Count elements greater than 0 in f_p_values
count_greater_than_zero = np.sum(chi_scores >= 1e+08)
print(f"Number of elements greater than 0: {count_greater_than_zero}")
```

```python
top_feature_count = 12

# Get feature names and create DataFrame for better analysis
feature_names = data.drop(['label'], axis=1).columns

# Create a DataFrame to map feature names to their chi2 scores and p-values
chi2_scores = f_p_values[0]
p_values = f_p_values[1]

# Create a DataFrame with feature information
feature_results = pd.DataFrame({
    'feature_name': feature_names,
    'chi2_score': chi2_scores,
    'p_value': p_values
})

# Remove features with NaN values
feature_results_clean = feature_results.dropna()

# Sort by chi2 score in descending order
feature_results_sorted = feature_results_clean.sort_values('chi2_score', ascending=False)

# Get top 8 features
top_features = feature_results_sorted.head(top_feature_count)

print(f"Top {top_feature_count} features by chi2 score:")
print(top_features[['feature_name', 'chi2_score', 'p_value']])

print(f"\nTop {top_feature_count} feature names:")
top_feature_names = top_features['feature_name'].tolist()
for i, name in enumerate(top_feature_names, 1):
    print(f"{i}. {name}")
```
